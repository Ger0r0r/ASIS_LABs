```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_excel('results.xlsx', header=None)
data = data.to_numpy()

x = data[0]
x = np.delete(x, 0)
print(x)
```

```python
styles = ['-', '--', '-.']
colors = ['red', 'blue', 'green', 'magenta','orange','brown']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

for i in range (12):
	if (i+3 != 14 and 
		i+3 != 11 and
		i+3 != 10 and
		i+3 != 9 and
		i+3 != 8):
			plt.plot(x, data[3+i,1:], label=data[3+i,0], color=colors[i % len(colors)],linestyle=styles[i // len(colors)], lw=2,)

# plt.title('Графики функций умножения в Python')
# plt.xlabel('Размер')
# plt.ylabel('Время')
# plt.ylim(0,10)

plt.legend()

plt.grid()
plt.show()
```
